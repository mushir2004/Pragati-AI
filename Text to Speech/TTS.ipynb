```python
import os
import io
from gtts import gTTS
import pdfplumber
from pptx import Presentation
import pytesseract
from PIL import Image
from google.colab import files
from IPython.display import display, Audio
import pytesseract
```

```python
# Function to extract text from different file types
def extract_text(file_path):
    file_extension = file_path.rsplit(".", 1)[-1].lower()
    text = ""

    if file_extension == "pdf":
        with pdfplumber.open(file_path) as pdf:
            for page in pdf.pages:
                page_text = page.extract_text()
                if page_text:
                    text += page_text + "\n"

    elif file_extension == "pptx":
        prs = Presentation(file_path)
        for slide in prs.slides:
            for shape in slide.shapes:
                if hasattr(shape, "text") and shape.text:
                    text += shape.text + "\n"

    elif file_extension in ["jpg", "jpeg", "png"]:
        image = Image.open(file_path)
        text = pytesseract.image_to_string(image)

    if not text.strip():
        raise ValueError(f"❌ No text found in {file_path}")

    return text.strip()

# Function to convert text to speech
def text_to_speech(text, output_file="output.mp3"):
    if not text.strip():
        raise ValueError("No text provided for conversion!")

    print(f"🔊 Converting text to speech...")
    tts = gTTS(text, lang="en")
    tts.save(output_file)
    print(f"✅ Speech saved as {output_file}")

    # 🎵 Play audio directly in Google Colab
    display(Audio(output_file, autoplay=True))

# User Selection: Text Input or File Upload
choice = input("📝 Choose input type (1: Text Input, 2: Upload File): ")

if choice == "1":
    user_text = input("📌 Enter text for conversion: ")
    if user_text.strip():
        text_to_speech(user_text)
    else:
        print("❌ No text entered!")

elif choice == "2":
    print("📂 Please upload your file (PDF, PPTX, JPG, PNG)")
    uploaded = files.upload()
    file_name = list(uploaded.keys())[0]  # Get uploaded file name

    # Save the file temporarily
    with open(file_name, "wb") as f:
        f.write(uploaded[file_name])

    print(f"✅ File uploaded: {file_name}")

    try:
        extracted_text = extract_text(file_name)
        text_to_speech(extracted_text)
    except Exception as e:
        print(f"❌ Error: {e}")
```
